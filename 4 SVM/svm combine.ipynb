```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python

# Load data from radiomics
data = pd.read_csv('RadiomicsFeatures.csv')
data.shape
```

```python
data2 = pd.read_csv('ResnetFeatures.csv')
data2.shape
```

```python
data2
```

```python
dfresnet = data2.loc[:,'0':'119']
dfresnet
```

```python
dfradio = data.loc[:,'original_shape_Elongation':'class']
dfradio
```

```python
print(dfresnet.std())
print(dfradio.std())
```

combining both dataframes

```python
final_df = pd.concat([dfresnet, dfradio], axis = 1)
```

```python
final_df
```

```python
x1 = final_df.loc[:,'0':'original_gldm_SmallDependenceLowGrayLevelEmphasis']
```

```python
x1
```

standardization

```python
x = np.asarray(x1)
y = np.asarray(final_df['class'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train_standardized = scaler.transform(x_train)
x_test_standardized = scaler.transform(x_test)
```

classification


```python
from sklearn import svm
classifier = svm.SVC(kernel = 'rbf', gamma = 'auto', C=2)
classifier.fit(x_train_standardized,y_train)
y_predict = classifier.predict(x_test_standardized)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn import svm
classifier = svm.SVC(kernel = 'linear', gamma = 'auto', C=2)
classifier.fit(x_train_standardized,y_train)
y_predict = classifier.predict(x_test_standardized)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

Evaluation

```python
correct = (y_test == y_predict).sum()
print("correct: ",correct)
wrong = x_test.shape[0] - correct
print("wrong: ",wrong)
```

```python
fraction_wrong = wrong/len(x_test)
print("Fraction classified incorrectly is: {:.2%}".format(fraction_wrong))
print("Accuracy: {:.2%}".format(1-fraction_wrong))
```
